The division algorithm states that for some integer $m$, there exists a unique quotient $q \in \mathbb{Z}$ and remainder $r \in \mathbb{Z}$ such that $m = qn + r$ for $0 \leq r < n$ where $n$ is some natural number. Letting $q = \lfloor \frac{m}{n} \rfloor$ and $r = m - qn$, it follows that $r \geq m - \left(\frac{m}{n} \right)n = 0$ and since $\frac{m}{n} < \lfloor \frac{m}{n} \rfloor + 1$, $r < m - \left(\frac{m}{n} - 1 \right)n = n$. Finally, uniqueness follows since $q_1n + r_1 = q_2n + r_2 \implies n$ divides $|r_1 - r_2| < n$ hence $r_1 = r_2$. Note the division algorithm can be extended for some non-negative integer $n$. 

By the Euclidean algorithm, we can find the gcd of $a$ and $b$ in polynomial steps in terms of the input size $\log(\min\{a, b\})$. The algorithm will terminate since the sequence $(a_i)$ is strictly decreasing. In the worst-case scenario, supposing the algorithm terminates after $i + 1$ steps, the quotients $q_0 = \dots = q_i = 1$. Then $r_j = r_{j + 1} + r_{j + 2}$ for $0 \leq j \leq i - 3$ hence $a_0 \geq F_{i + 1}$ and $b_0 \geq F_{i + 2}$ where $F_n$ is the $n^{\text{th}}$ Fibonnaci number starting with $F_1 = 1$ and $F_2 = 1$. By Binet's formula, $F_{i + 2} = \Theta(\phi^{i + 2})$ thus, since we have $\log(b_0) \geq \log(F_{i + 2})$, $i = O(\log(b_0))$. 

Running the Extended Euclidean algorithm gives Bezout's lemma, which states for integers $m$ and $n$, there exists integers $x$ and $y$ such that $mx + ny = (x, y)$. 

```python
import math


a, b = 39, 35 
row_1, row_2 = [1, 0, b], [0, 1, a]

while b > 0:
    r = a % b 
    a, b = b, r

print(f"{row_1[2]}x + {row_2[2]}y = {a}")
print("\nrow_1, row_2:")

while row_2[2] > 0:
    rem = math.floor(row_1[2] / row_2[2])
    row_3 = [row_1[i] - rem * row_2[i] for i in range(3)]
    row_1, row_2 = row_2, row_3
    print(row_1, row_2)

print("\nsolution: ")
print(f"x = {row_1[0]} + {row_2[0]}k")
print(f"y = {row_1[1]} + {row_2[1]}k")
```

To understand the algorithm, it's helpful to do the Euclidean algorithm first:

$$
\begin{align*}
    39 & = 1 \times 35 + 4 \\
    35 & = 8 \times 4 + 3 \\
    4 & = 1 \times 3 + 1 \\
    3 & = 3 \times 1 + 0
\end{align*}
$$

We begin with the equations $35 \times 0 + 39 \times 1 = 39 (1)$ and $35 \times 1 + 0 \times 39 = 35 (2)$ and apply repeated subtractions to reduce the right-hand side of (1) until it equals $\mathrm{gcd}(39, 35) = 1$. 

Note that `row_1` initially represents equation (1) but we swap `row_1` and `row_2` at the first step so the larger number, 39, appears as the right-hand side of `row_1`. At each iteration, we subtract $\lfloor \frac{u_1}{u_2} \rfloor$ copies of `row_2` from `row_1`, where $u_i$ denotes the third element of `row_i`. This reduces the right-hand side of equation (1) to the remainder. For example, since $\lfloor \frac{39}{35} = 1$, we compute $(1) - 1 \times (2)$ to obtain $-1 \times 35 + 1 \times 39 = 4$. 

Finally, we obtain the general solution by noting if $(x_0, y_0)$ solves $35x_0 + 39y_0 = 1$ then $35(x_0 + 39k) + 39(y_0 - 35k) = 1$ for some integer $k$.
